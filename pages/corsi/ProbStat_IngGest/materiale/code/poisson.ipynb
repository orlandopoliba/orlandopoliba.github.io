```python
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
n = 100
l = 1.2
p = l/n

binomial_value = np.zeros(n+1)
for k in range(n+1):
    binomial_value[k] = math.factorial(n)/(math.factorial(k)*math.factorial(n-k))*(p**k)*((1-p)**(n-k))    
```

```python
fig, ax = plt.subplots(1, 1)
ax.plot(range(20),binomial_value[0:20], 'bo', ms=8)
```

```python
poisson_value = np.zeros(n+1)
for k in range(n+1):
    poisson_value[k] = l**k/math.factorial(k)*math.exp(-l)
```

```python
fig, ax = plt.subplots(1, 1)
ax.plot(range(20),binomial_value[0:20], 'bo', ms=8)
ax.plot(range(20),poisson_value[0:20], 'ro', ms=8)
```

```python
binomial_sample = np.random.binomial(n, p, 1000)
```

```python
frequencies = np.zeros(n+1)
for k in range(n+1):
    frequencies[k] = binomial_sample.tolist().count(k)/1000
    
fig, ax = plt.subplots(1, 1)
ax.plot(range(20),frequencies[0:20], 'bo', ms=8)
```

```python
poisson_sample = np.random.poisson(l, 1000)
```

```python
frequencies = np.zeros(n+1)
for k in range(n+1):
    frequencies[k] = poisson_sample.tolist().count(k)/1000
    
fig, ax = plt.subplots(1, 1)
ax.plot(range(20),frequencies[0:20], 'bo', ms=8)
```
