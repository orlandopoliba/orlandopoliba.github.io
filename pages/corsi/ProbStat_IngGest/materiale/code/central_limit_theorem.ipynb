```python
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
def f(x,mu,sigmaSquared): 
    sigma = np.sqrt(sigmaSquared)
    return 1/(sigma*np.sqrt(2*np.pi))*np.exp(-1/2*((x-mu)/sigma)**2)
```

```python
def CLT_Z(X,mu,sigmaSquared):
    n = len(X)
    sigma = np.sqrt(sigmaSquared)
    return (1/n*np.sum(X) - mu)/(sigma/np.sqrt(n))
```

```python
# binomial

n = 20
p = 1/3

mu = n*p
sigmaSquared = n*p*(1-p)

# ripetiamo N volte l'esperimento

N = 1000 
Z = np.zeros(N)

K = 30 # numero 'grande' di dati estratti da X: l'n del TLC

for i in range(N):
    X = np.random.binomial(n,p,K)
    Z[i] = CLT_Z(X,mu,sigmaSquared)

# plot

bins = np.linspace(-4,4,50)
x = np.linspace(-4,4,100)

fig, ax = plt.subplots()
n, bins, patches = ax.hist(Z, bins, rwidth=0.85, density=True)
ax.plot(x,f(x,0,1))
ax.grid(axis='y', alpha=0.75)
plt.title('binomial')  
```

```python
# poisson

l = 1.2

mu = l
sigmaSquared = l

# estraiamo un campione di ampiezza N della v.a. usata nel TLC

N = 1000
Z = np.zeros(N)

K = 30 # numero 'grande' di dati estratti da X: l'n del TLC

for i in range(N):
    X = np.random.poisson(l,K)
    Z[i] = CLT_Z(X,mu,sigmaSquared)


bins = np.linspace(-4,4,50)
x = np.linspace(-4,4,100)

fig, ax = plt.subplots()
n, bins, patches = ax.hist(Z, bins, rwidth=0.85, density=True)
ax.plot(x,f(x,0,1))
ax.grid(axis='y', alpha=0.75)
plt.title('poisson')  
```

```python
# geometrica

p = 1/5

mu = 1/p
sigmaSquared = (1-p)/p**2

# estraiamo un campione di ampiezza N della v.a. usata nel TLC

N = 1000
Z = np.zeros(N)

K = 30 # numero 'grande' di dati estratti da X: l'n del TLC

for i in range(N):
    X = np.random.geometric(p,K)
    Z[i] = CLT_Z(X,mu,sigmaSquared)


bins = np.linspace(-4,4,50)
x = np.linspace(-4,4,100)

fig, ax = plt.subplots()
n, bins, patches = ax.hist(Z, bins, rwidth=0.85, density=True)
ax.plot(x,f(x,0,1))
ax.grid(axis='y', alpha=0.75)
plt.title('geometric')  
```

```python
# gamma

alpha = 11/2
l = 1/2

mu = alpha/l
sigmaSquared = alpha/l**2

# estraiamo un campione di ampiezza N della v.a. usata nel TLC

N = 1000
Z = np.zeros(N)

K = 30 # numero 'grande' di dati estratti da X: l'n del TLC

for i in range(N):
    X = np.random.gamma(alpha, 1/l ,K)
    Z[i] = CLT_Z(X,mu,sigmaSquared)


bins = np.linspace(-4,4,50)
x = np.linspace(-4,4,100)

fig, ax = plt.subplots()
n, bins, patches = ax.hist(Z, bins, rwidth=0.85, density=True)
ax.plot(x,f(x,0,1))
ax.grid(axis='y', alpha=0.75)
plt.title('gamma')  
```
