```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-1.0, 1.0, 100)
y = np.linspace(-1.0, 1.0, 100)
X, Y = np.meshgrid(x,y)


plt.gca().set_aspect('equal')

number_of_sampled_points = 1000
number_of_points_inside_circle = 0



for i in range(0,number_of_sampled_points):
    P = np.random.uniform(-1,1,size=2)
    if np.dot(P,P) <= 1:
        number_of_points_inside_circle = number_of_points_inside_circle + 1
        plt.plot([P[0]],[P[1]],'.', color='blue',alpha=0.5)
    else:
        plt.plot([P[0]],[P[1]],'.', color='red',alpha=0.5)

F = X**2 + Y**2 - 1
plt.contour(X,Y,F,[0])
plt.show()

print('Frazione di punti caduti nel cerchio per 4: ', number_of_points_inside_circle/number_of_sampled_points*4, sep = '')
print('pi greco: ', np.pi)
```

```python
# faster code

from numba import jit

@jit(nopython=True)
def large_numbers(number_of_sampled_points):
    number_of_points_inside_circle = 0
    for i in range(0,number_of_sampled_points):
        P = np.random.uniform(-1,1,size=2)
        if np.dot(P,P) <= 1:
            number_of_points_inside_circle = number_of_points_inside_circle + 1
    return number_of_points_inside_circle/number_of_sampled_points

p = large_numbers(100_000_000)        

print('Frazione di punti caduti nel cerchio per 4: ', p*4, sep = '')
print('pi greco: ', np.pi)
```
