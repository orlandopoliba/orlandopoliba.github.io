```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as patches
```

```python
plt.gca().set_aspect('equal')

for i in range(0,1000):
    theta1 = np.random.uniform(0,2*np.pi)
    p1 = [np.cos(theta1), np.sin(theta1)]
    theta2 = np.random.uniform(0,2*np.pi)
    p2 = [np.cos(theta2), np.sin(theta2)]
    plt.plot([p1[0],p2[0]],[p1[1],p2[1]],'0.6', linewidth=0.2)

x = np.linspace(-1.0, 1.0, 100)
y = np.linspace(-1.0, 1.0, 100)
X, Y = np.meshgrid(x,y)
F = X**2 + Y**2 - 1
plt.contour(X,Y,F,[0])
plt.show()
```

```python
x = np.linspace(-1.0, 1.0, 100)
y = np.linspace(-1.0, 1.0, 100)
X, Y = np.meshgrid(x,y)


plt.gca().set_aspect('equal')

for i in range(0,1000):
    theta = np.random.uniform(0,2*np.pi)
    r = np.random.uniform(0,1)

    R = np.cos(theta)*(X-r*np.cos(theta)) + np.sin(theta)*(Y-r*np.sin(theta))
    plt.contour(X,Y,R,[0], colors='0.6', linewidths=0.2)

F = X**2 + Y**2 - 1
plt.contour(X,Y,F,[0])
plt.show()
```

```python
x = np.linspace(-1.0, 1.0, 100)
y = np.linspace(-1.0, 1.0, 100)
X, Y = np.meshgrid(x,y)


plt.gca().set_aspect('equal')

for i in range(0,1000):
    theta = np.random.uniform(0,2*np.pi)
    r = np.sqrt(np.random.uniform(0,1))

    R = np.cos(theta)*(X-r*np.cos(theta)) + np.sin(theta)*(Y-r*np.sin(theta))
    plt.contour(X,Y,R,[0], colors='0.6', linewidths=0.2)

F = X**2 + Y**2 - 1
plt.contour(X,Y,F,[0])
plt.show()
```
